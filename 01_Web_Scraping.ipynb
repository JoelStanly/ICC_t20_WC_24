# Importing selenium for webDriver to fetch the scorecard and other details

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
def extract_wickets(score):
    if "/" in score:
        return int(score.split("/")[1])
    else:
        return 10
```

```python
def scorecard_fetch(url,match_id,teams_list, battingSummary_list, bowlingSummary_list, count):
    battingSummary = {}
    battingSummary['matchId'] = match_id
    battingSummary["battingScore"] = []
    bowlingSummary = {}
    bowlingSummary['matchId'] = match_id
    bowlingSummary["bowlingInfo"] = []
    
    driver.get(url)
    time.sleep(2)
    action = ActionChains(driver)
    driver.implicitly_wait(5)
    scorecard_button = driver.find_element(By.CSS_SELECTOR,'li[role="presentation"] button[aria-controls="scorecard"]')
    driver.execute_script("arguments[0].scrollIntoView();", scorecard_button)
    action.move_to_element(scorecard_button).click().perform()
    # fetch the scorecard for the particular innings
    print(len(driver.find_element(By.ID, "scorecard-widget")
                   .find_element(By.CLASS_NAME, "si-tabs")
                   .find_elements(By.TAG_NAME, "li")))
    driver.implicitly_wait(5)
    
    for i in range(len(driver.find_element(By.ID, "scorecard-widget")
                   .find_element(By.CLASS_NAME, "si-tabs")
                   .find_elements(By.TAG_NAME, "li"))):
        scorecard_widget = driver.find_element(By.ID, "scorecard-widget")
        si_tabs_ul = scorecard_widget.find_element(By.CLASS_NAME, "si-tabs")
        list_items = si_tabs_ul.find_elements(By.TAG_NAME, "li")
        print(i)
        items = list_items[i]
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "si-scorecard-wrapper"))
        )
        WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable(items)
        )
        driver.execute_script("arguments[0].click();", items)
        #action.move_to_element(items).click().perform()
        batting_team = items.find_element(By.TAG_NAME,"span").text[:3].strip()
        si_tab_container = scorecard_widget.find_element(By.CLASS_NAME, "si-tab-container")
        scorecard_wrapper = si_tab_container.find_element(By.CLASS_NAME,"si-tab-container-item").find_element(By.CLASS_NAME,"si-scorecard-wrapper")
        score_detail = scorecard_wrapper.find_element(By.CLASS_NAME, "si-team-score-detail")
        si_team = score_detail.find_element(By.CLASS_NAME,"si-team")
        batting_team_flag = si_team.find_element(By.CLASS_NAME, "si-team-flag")
        si_score = score_detail.find_element(By.CLASS_NAME,"si-score")
        si_batting_scorecard = scorecard_wrapper.find_element(By.CLASS_NAME,"si-batting-scorecard")
        si_batting_table = si_batting_scorecard.find_element(By.CLASS_NAME,"si-batting-table")
        si_table_body = WebDriverWait(si_batting_table, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "si-table-body"))
        )
        #si_table_body = si_batting_table.find_element(By.CLASS_NAME, "si-table-body")
        si_table_rows = si_table_body.find_elements(By.XPATH, "./div[contains(@class, 'si-table-row')]")
        si_extras_scorecard = scorecard_wrapper.find_element(By.CLASS_NAME, "si-extras-scorecard")
        si_teams_scorecard = scorecard_wrapper.find_element(By.CLASS_NAME, "si-teams-scorecard")
        si_bowling_scorecard = si_teams_scorecard.find_element(By.CLASS_NAME, "si-bowling-scorecard")
        si_team_name = si_bowling_scorecard.find_element(By.CLASS_NAME,"si-team-name")
        si_bowling_table = si_teams_scorecard.find_element(By.CLASS_NAME,"si-bowling-table")
        si_bowling_table_body = WebDriverWait(si_bowling_table, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "si-table-body"))
        )
        #si_bowling_table_body = si_bowling_table.find_element(By.CLASS_NAME,"si-table-body")
        si_bowling_table_row = si_bowling_table_body.find_elements(By.XPATH, "./div[contains(@class, 'si-table-row')]")
        # Team dict values fetching
        teams = {}
        teams["team"] = batting_team
        teams["flag"] = batting_team_flag.find_element(By.TAG_NAME,"img").get_attribute("src")
        teams["teamName"] = si_team.find_element(By.TAG_NAME,"span").text.strip()
        # Batting Summary Values Fetching
        batting_score = {}
        batting_score["battingTeam"] = batting_team
        batting_score["CRR"] = si_score.find_element(By.CLASS_NAME,"over-runrate").text.split()[1].strip()
        batting_score["overs"] = si_score.find_element(By.CLASS_NAME,"overs").text.strip("()").split()[0].strip()
        batting_score["totalScore"] = si_score.find_element(By.CLASS_NAME,"runs").text.strip().split("/")[0].strip()
        batting_score["wickets"] = extract_wickets(si_score.find_element(By.CLASS_NAME,"runs").text.strip())
        batting_score["scores"] = []
        players = []
        for rows in si_table_rows:
            print("Processing a row bats")
            player_score = {}
            si_accordion_head = rows.find_element(By.CLASS_NAME,"si-accordion-head")
            si_player_table_row = si_accordion_head.find_element(By.CLASS_NAME,"si-table-row")
            si_player_wrap = si_player_table_row.find_element(By.CLASS_NAME,"si-label").find_element(By.CLASS_NAME,"si-player-wrap")
            player_score["playerName"] = si_player_wrap.find_element(By.CLASS_NAME,"si-player-info").find_element(By.TAG_NAME,"a").find_element(By.CLASS_NAME,"si-name").text.strip()
            player_score["status"] = si_player_wrap.find_element(By.CLASS_NAME,"si-player-info").find_element(By.CLASS_NAME,"si-status").text.strip()
            player_score["runs"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-runs").find_element(By.TAG_NAME,"span").text.strip().replace("-","")
            player_score["balls"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-balls").find_element(By.TAG_NAME,"span").text.strip().replace("-","")
            player_score["fours"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-fours").find_element(By.TAG_NAME,"span").text.strip().replace("-","")
            player_score["sixes"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-sixes").find_element(By.TAG_NAME,"span").text.strip().replace("-","")
            player_score["strikeRate"] = si_player_table_row.find_element(By.CLASS_NAME,"si-strike-rate").find_element(By.TAG_NAME,"span").text.strip().replace("-","")
            player = {}
            player["playerName"] = player_score["playerName"]
            player["playerImage"] = si_player_wrap.find_element(By.TAG_NAME,"img").get_attribute("src")
            player["playerstats"] = si_player_wrap.find_element(By.CLASS_NAME,"si-player-info").find_element(By.TAG_NAME,"a").get_attribute("href")
            
            players.append(player)
            batting_score["scores"].append(player_score)

        #Batting Loop end
        teams["players"] = players
        battingSummary["battingScore"].append(batting_score)

        # Bowling Summary
        extras = {}
        extras_list = si_extras_scorecard.find_element(By.CLASS_NAME,"si-info").text.strip().split()
        for extra in extras_list:
            value = int(''.join(filter(str.isdigit, extra)))
            key = ''.join(filter(str.isalpha, extra))
            extras[key] = value
        total_extras = int(si_extras_scorecard.find_element(By.CLASS_NAME,"si-total").text.strip())
        bowling_info = {}
        bowling_info["bowlingTeam"] = si_team_name.text.strip()
        bowling_info["extras"] = extras
        bowling_info["extrasTotal"] = total_extras
        bowling_info["bowling"] = []
        for rows in si_bowling_table_row:
            print("Processing a row bowl")
            bowlData = {}
            si_accordion_head = rows.find_element(By.CLASS_NAME,"si-accordion-head")
            si_player_table_row = si_accordion_head.find_element(By.CLASS_NAME,"si-table-row")
            si_player_wrap = si_player_table_row.find_element(By.CLASS_NAME,"si-label").find_element(By.CLASS_NAME,"si-player-wrap")
            bowlData["playerName"] = si_player_wrap.find_element(By.CLASS_NAME,"si-player-info").find_element(By.TAG_NAME,"a").find_element(By.CLASS_NAME,"si-name").text.strip()
            bowlData["overs"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-overs").find_element(By.TAG_NAME,"span").text.strip()
            bowlData["runs"] = si_player_table_row.find_element(By.CLASS_NAME,"si-player-runs").find_element(By.TAG_NAME,"span").text.strip()
            si_player_wickets = si_player_table_row.find_elements(By.CLASS_NAME,"si-player-wickets")
            bowlData["maidens"] = si_player_wickets[0].find_element(By.TAG_NAME,"span").text.strip()
            bowlData["wickets"] = si_player_wickets[1].find_element(By.TAG_NAME,"span").text.strip()
            bowlData["economy"] = si_player_wickets[2].find_element(By.TAG_NAME,"span").text.strip()
            bowling_info["bowling"].append(bowlData)

        # Bowling Loop end
        bowlingSummary["bowlingInfo"].append(bowling_info)
        if not any(team["team"] == teams["team"] for team in teams_list):
            teams_list.append(teams)
    battingSummary_list.append(battingSummary)
    bowlingSummary_list.append(bowlingSummary)
    if scorecard_button.get_attribute('aria-selected') == 'true':
        return teams_list, battingSummary_list, bowlingSummary_list, count+ 1
    return scorecard_fetch(url,match_id,teams_list, battingSummary_list, bowlingSummary_list, count)
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
count = 0
battingSummary_list = []
bowlingSummary_list = []
teams_list = []
for index,row in df_matches.iterrows():
    match_id = row['matchId']
    matching_match = next((match for match in match_detail_info if match['matchId'] == match_id), None)
    teams_list, battingSummary_list, bowlingSummary_list, count = scorecard_fetch(matching_match['matchInfoURL'],match_id,teams_list, battingSummary_list, bowlingSummary_list, count)
    print(str(count) +" matches fetched")
    #break
print("Fetch completed")
driver.quit()
```

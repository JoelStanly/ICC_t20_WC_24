# Cleaning the match Result and transforming -> margin

```python
def matchResultCleaner(matchResult):
    margin_pattern = r"(\d+)\s(runs|wickets)"
    super_over_pattern = r"Super Over by (\d+)\s(?:runs)"
    if "Super Over" in matchResult:
        match = re.search(super_over_pattern,matchResult)
        if match:
            return f"{match.group(1)} runs (Super Over)"
    else:
        match = re.search(margin_pattern,matchResult)
        if match:
            return f"{match.group(1)} {match.group(2)}"
    return "No Margin"
matchResultCleaner("USA beat Canada by 7 wickets")
```

```python
df_matches['margin'] = df_matches['matchResult'].apply(matchResultCleaner)
df_matches.head(10)
```

# Flatten method for flattening the data to able for dataframe transformation

```python
def flatten_method(unflat_list):
    flattened_data = []
    for item in unflat_list:
        temp_data = {}
        for key, value in item.items():
            if isinstance(value, str):
                temp_data[key] = value
            if isinstance(value, dict):
                for k,v in value.items():
                    temp_data[k] = v
        for key, value in item.items():
            if isinstance(value, list):
                for list_item in value:
                    list_item.update(temp_data)
                flattened_data.append(value)
    flattened_data = [item for sublist in flattened_data for item in sublist]
    return flattened_data
```
